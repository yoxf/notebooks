# Sample plots in Matplotlib

Line Plot  折线图
=========

函数:`~matplotlib.pyplot.plot`可以用来创建折线图
```
plot(x, y)        # plot x and y using default line style and color
plot(x, y, 'bo')  # plot x and y using blue circle markers
plot(y)           # plot y using x as index array 0..N-1
plot(y, 'r+')     # ditto, but with red plusses
```


#### Pyplot Simple

plot(x,y,*args)
> x  横坐标
>
> y  纵坐标

````python
import matplotlib.pyplot as plt
plt.style.use('ggplot')
fig= plt.figure(figsize = (8,4))
ax = fig.add_subplot(111)
x= [1,2,3,4]
y =[1,6,0.3,9]
ax.plot(x,y,"ro-") 
ax.set_ylabel('some numbers')
ax.grid(True)
# 设置y轴坐标范围
ax.set_ylim(0,20,10)
# 设置数据标签
for a,b in zip(x,y):
    plt.text(a, b+0.5, '%.0f' % b, ha='center', va= 'bottom',fontsize=10,color='r')
plt.show()
````

#### Pyplot Three

````python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('ggplot')
# evenly sampled time at 200ms intervals
t = np.arange(0., 5., 0.2)

# red dashes, blue squares and green triangles
fig = plt.figure(figsize = (8,4))
ax = fig.add_subplot(111)
ax.plot(t, t, 'r--', t, t**2, 'bs-', t, t**3, 'g^-')
plt.show()
````

#### Stackplot 堆积面积图

Stackplot(x,y1,y2,...,)
> x:横坐标
>
> y1: 纵坐标1
> 
> y2:纵坐标2
> 
> y3:纵坐标3
> 

````python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('ggplot')
x = [1, 2, 3, 4, 5]  # x坐标
y1 = [3, 5, 2, 3, 5] 
y2 = [0, 4, 2, 6, 8]
y3 = [1, 3, 5, 7, 5]
labels = ["Fibonacci ", "Evens", "Odds"]  #图例标签

#fig, ax = plt.subplots() # fig,ax写法可以同时建立画布和坐标系
fig =plt.figure(figsize = (8,4))
ax = fig.add_subplot(111)
ax.stackplot(x, y1, y2, y3, labels=labels)
ax.legend(loc=2)
plt.show()
````

Bar charts 条形图
==========

函数`~matplotlib.pyplot.bar`用来制作条形图（包括定制化的图形如误差条形图，堆积条形图，水平条形图等）
```
bar(x, height, *, align='center', **kwargs)
bar(x, height, width, *, align='center', **kwargs)
bar(x, height, width, bottom, *, align='center', **kwargs)
```

#### Bar graphs

````python
import numpy as np
import matplotlib.pyplot as plt

# data
x = np.arange(5)
y1 = [21, 13, 21, 14,  7]
y2 = [ 3,  3, 24,  4, 18]
width = 0.25
# plot
fig = plt.figure(figsize = (8,4))
ax = fig.add_subplot(111)
ax.bar(x, y1, width,color ='r')
ax.bar(x + width, y2, width,color='k')
# setting
ax.set_title("Bar graphs")
ax.set_xticks(x + width,['a', 'b', 'c', 'd', 'e']) #没有显示X刻度
# 设置y轴范围
ax.set_ylim(0,30)
# 设置数据标签
for a,b,c in zip(x,y1,y2):
    plt.text(a, b+0.5, '%.0f' % b, ha='center', va= 'bottom',fontsize=10,color='r')
    plt.text(a + 0.25, c+0.5, '%.0f' % c, ha='center', va= 'bottom',fontsize=10,color='k')

plt.show()
````

#### Stacked Bar Graph

````python
import numpy as np
import matplotlib.pyplot as plt
# data
N = 5
menMeans = (20, 35, 30, 35, 27)  
womenMeans = (25, 32, 34, 20, 25)
menStd = (2, 3, 4, 1, 2)
womenStd = (3, 5, 2, 3, 3)
ind = np.arange(N)    # the x locations for the groups
width = 0.35       # the width of the bars: can also be len(x) sequence

# plot
fig = plt.figure(figsize= (8,4))
ax = fig.add_subplot(111)
# x索引（坐标相同时），产生堆积条形图
# yerr产生误差线
p1 = ax.bar(ind, menMeans, width, yerr=menStd,color = '#006633')
p2 = ax.bar(ind, womenMeans, width,bottom=menMeans, yerr=womenStd,color = 'pink')

# setting
ax.set_ylabel('Scores') # y轴标题
ax.set_title('Scores by group and gender') #图表标题
ax.set_xticks(ind, ('G1', 'G2', 'G3', 'G4', 'G5')) # x轴刻度
ax.set_yticks(np.arange(0, 81, 20)) # 刻度，最后一位是刻度大小
ax.legend((p1[0], p2[0]), ('Men', 'Women')) # 图例
menMeans = (20, 35, 30, 35, 27)  
womenMeans = (25, 32, 34, 20, 25)

# 设置y轴范围
ax.set_ylim(0,85)
# 设置数据标签
for a,b,c in zip(ind,menMeans,womenMeans):
    plt.text(a, b+0.5, '%.0f' % b, ha='center', va= 'bottom',fontsize=10,color='#006633')
    plt.text(a, b+c+5.5, '%.0f' % c, ha='center', va= 'top',fontsize=10,color='r')

plt.show()
````

#### Horizontal bar chart

````python
import matplotlib.pyplot as plt
import numpy as np

plt.style.use('ggplot')
# Fixing random state for reproducibility
np.random.seed(19680801)
# Example data
people = ('Tom', 'Dick', 'Harry', 'Slim', 'Jim')  #y坐标标签
y_pos = np.arange(len(people)) 
performance = [ 10.,  10.42,  10.09,   8.66,  12.77]
error = [ 0.70 ,  0.74,  0.70,  0.56,  0.97]
width = 0.45

#plot
fig = plt.figure(figsize = (8,4))
ax = fig.add_subplot(111)
ax.barh(y_pos, performance, width,xerr=error, align='center',color= '#159FF8',ecolor= 'k')
# setting
ax.set_yticks(y_pos,people)
ax.set_xlabel('Performance')
ax.set_title('How fast do you want to go today?')

# 设置y轴范围
ax.set_xlim(0,20)
# 设置数据标签
for a,b in zip(performance,y_pos):
    plt.text(a+0.5, b, '%0.00f' % a, ha='center', va= 'bottom',fontsize=12,color='#159FF8')

plt.show()
````

#### sample bar plot on a polar axis 简单玫瑰图

````python
import numpy as np
import matplotlib.pyplot as plt
# 似乎宽度要设定成一个定值，只有半径变化，数值大小通过面积来反映。
# 用途：在条形图或堆积条形图不能直观对比某此变更后数据整体的变化时（如某种药物的使用效果，广告投放效果等）
np.random.seed(19680801)
# Compute pie slices
N = 6  # 玫瑰花瓣个数
theta = np.linspace(0.0, 2 * np.pi, N, endpoint=False)  # 角度
radii = [ 96,  121,  100,
        111, 102,  81]
width = [ 0.5, 0.5,  0.5, 
        0.5,  0.5,  0.5]
plt.subplot(111, projection='polar') # 极坐标必须这样写
bars = plt.bar(theta, radii, width=width, bottom=0.0)  #条形图绘制

plt.show()
````

#### bar plot on a polar axis 玫瑰图(极坐标条形图)

````python
import numpy as np
import matplotlib.pyplot as plt

# Fixing random state for reproducibility
np.random.seed(19680801)

# Compute pie slices
N = 20  # 玫瑰花瓣个数
theta = np.linspace(0.0, 2 * np.pi, N, endpoint=False)  # 角度
radii = 10 * np.random.rand(N)  #半径
width = np.pi / 4 * np.random.rand(N) # 宽度

fig = plt.figure(figsize = (8,8))
ax = fig.add_subplot(221)
plt.subplot(221, projection='polar') # 极坐标必须这样写
bars = plt.bar(theta, radii, width=width, bottom=0.0)  #条形图绘制

ax = ax = fig.add_subplot(222)
plt.subplot(222, projection='polar') # 极坐标必须这样写
bars = plt.bar(theta, radii*2, width=width*1.5, bottom=0.0)  #条形图绘制

# 使用自定义颜色
for r, bar in zip(radii, bars):
    bar.set_facecolor(plt.cm.viridis(r / 10.))
    bar.set_alpha(1)

plt.show()
````

#### Basic pie chart

Scatter plots 散点图
=============

函数`~matplotlib.pyplot.scatter` 用来制作散点图

```
matplotlib.pyplot.scatter(x, y, s=None, c=None, marker=None, cmap=None, norm=None, vmin=None, vmax=None, alpha=None, linewidths=None, verts=None, edgecolors=None, hold=None, data=None, **kwargs)
Make a scatter plot of x vs y.

Marker size is scaled by s and marker color is mapped to c.
```

#### Scatter plot

````python
import numpy as np
import matplotlib.pyplot as plt

# Fixing random state for reproducibility
np.random.seed(19680801)
x = [2,4,5,7,
     12,19,23,54]
y = [6,22,1,19,
     22,18,12,9]
#colors = ['r','k','b','g','y','g','b','r']
area = [230,1000,102,400,
        33,240,90,252]
colors = np.sqrt(area)  #通过公式造颜色序列
#plot
fig = plt.figure(figsize = (8,4))
ax = fig.add_subplot(111)
ax.scatter(x, y, s=area, c=colors, alpha=0.5)

for a,b,c in zip(x,y):
    plt.text(a, b+0.5, '%.0f' % b, ha='center', va= 'bottom',fontsize=10,color='r')
    plt.text(a + 0.25, c+0.5, '%.0f' % c, ha='center', va= 'bottom',fontsize=10,color='k')
plt.show()
````

#### Scatter Masked

````python
import matplotlib.pyplot as plt
import numpy as np

# Fixing random state for reproducibility
np.random.seed(19680801)

#data
N = 100
r0 = 0.6
x = 0.9 * np.random.rand(N)
y = 0.9 * np.random.rand(N)
area = np.pi * (10 * np.random.rand(N))**2  # 0 to 10 point radii
c = np.sqrt(area)
r = np.sqrt(x * x + y * y)
area1 = np.ma.masked_where(r < r0, area)
area2 = np.ma.masked_where(r >= r0, area)

# plot
fig = plt.figure(figsize = (8,4))
ax = fig.add_subplot(111)
ax.scatter(x, x+y, s=area1, marker='^',c=c)
ax.scatter(x, y, s=area2, marker='o',c=c)
# Show the boundary between the regions:
theta = np.arange(0, np.pi / 2, 0.01)
plt.plot(r0 * np.cos(theta), r0 * np.sin(theta))

plt.show()
````

#### Scatter Star Poly

````python
import numpy as np
import matplotlib.pyplot as plt

# Fixing random state for reproducibility
np.random.seed(19680801)

#data
x = np.random.rand(10)
y = np.random.rand(10)
z = np.sqrt(x**2 + y**2)
#plot
fig = plt.figure(figsize=(12,10))
plt.subplot(321)
plt.scatter(x, y, s=80, c=z, marker=">")

plt.subplot(322)
plt.scatter(x, y, s=80, c=z, marker=(5, 0))

verts = list(zip([-1., 1., 1., -1.], [-1., -1., 1., -1.]))
plt.subplot(323)
plt.scatter(x, y, s=80, c=z, marker=(verts, 0))
# equivalent:
# plt.scatter(x, y, s=80, c=z, marker=None, verts=verts)

plt.subplot(324)
plt.scatter(x, y, s=80, c=z, marker=(5, 1))

plt.subplot(325)
plt.scatter(x, y, s=80, c=z, marker='+')

plt.subplot(326)
plt.scatter(x, y, s=80, c=z, marker=(5, 2))

plt.show()
````

### Sample radar chart 简单雷达图

````python
import numpy as np
import matplotlib.pyplot as plt
#plt.style.use("ggplot")
#标签
labels = np.array(['艺术A','调研I','实际R','常规C','企业E','社会S'])
#数据个数
dataLenth = 6
#数据
data = np.array([1,4,3,6,4,8])


angles = np.linspace(0, 2*np.pi, dataLenth, endpoint=False)
data = np.concatenate((data, [data[0]])) # 闭合
angles = np.concatenate((angles, [angles[0]])) # 闭合

fig = plt.figure()
plt.subplot(111, polar=True)# polar参数！！
plt.plot(angles, data, 'ro-', linewidth=2)# 画线
plt.fill(angles, data, facecolor='r', alpha=0.25)# 填充
plt.thetagrids(angles * 180/np.pi, labels, fontproperties="SimHei")
plt.title("matplotlib雷达图", va='bottom', fontproperties="SimHei")
plt.grid(True)

plt.show()
````

#### 组分复杂的数据占比分析

````python
# 导入第三方包
%matplotlib inline
import matplotlib.pyplot as plt
import squarify

# r style
plt.style.use("ggplot")
#中文及负号处理办法
plt.rcParams['font.sans-serif'] = 'Microsoft YaHei'
plt.rcParams['axes.unicode_minus'] = False

# 创建数据
name = ['国内增值税','国内消费税',
        '企业所得税','个人所得税',
        '进口增值税,消费税','出口退税',
        '城市维护建设税', '车辆购置税',
        '印花税','资源税',
        '土地和房税','车船税烟叶税等']
income = [3908,856,801,868,1361,1042,320,291,175,111,414,363]
        
# 绘图
#colors = ['b','r','red','indianred', 'green','pink','orange']
fig = plt.figure(figsize = (10,6))
ax = fig.add_subplot(111)
plot = squarify.plot(sizes = income, # 指定绘图数据
                     label = name, # 指定标签
                     #color = colors, # 指定自定义颜色
                     alpha = 0.9, # 指定透明度
                     value = income, # 添加数值标签
                     edgecolor = 'white', # 设置边界框为白色
                     linewidth =0.1 # 设置边框宽度为3
                    )
# 设置标签大小
plt.rc('font', size=4)
# 设置标题大小
plot.set_title('2017年8月中央财政收支情况',
               fontsize = 20,
               #fontdict = {'fontsize':15}
              )
# 去除坐标轴
plt.axis('off')
# 去除上边框和右边框刻度
plt.tick_params(top = 'off', right = 'off')
# 显示图形
plt.show()
````

# Map plot 地图

### Map plot  地图

````python
%matplotlib inline
import cartopy.crs as ccrs
import matplotlib.pyplot as plt
# r style
plt.style.use("ggplot")
#中文及负号处理办法
plt.rcParams['font.sans-serif'] = 'Microsoft YaHei'
plt.rcParams['axes.unicode_minus'] = False

fig = plt.figure(figsize=(20,10))
ax = plt.axes(projection=ccrs.PlateCarree())
ax.coastlines()  # 绘制海岸线
#ax = plt.axes(projection=ccrs.Mollweide())
ax.stock_img()  #加载背景图

# 设定经纬度坐标
ny_lon, ny_lat = -75, 43    # Longitude 经度，Latitude 纬度 （纽约）
delhi_lon, delhi_lat = 77.23, 28.61  # 新德里
beij_lon, beij_lat =116.40,39.90 # 北京
gz_lon,gz_lat =113.264,23.131 # 广州

# transform把坐标xdata,ydata 转换到球面（仿射转换）
plt.plot([ny_lon, beij_lon], [ny_lat,beij_lat], # xdata,ydata
         color='blue',
         linewidth=1, 
         marker='o',
         #transform=ccrs.Geodetic(), #定义了一个纬度/经度坐标系与球面拓扑, 地理距离和坐标测量的程度。
          transform= ccrs.PlateCarree()
         )
plt.plot([ny_lon, delhi_lon], [ny_lat, delhi_lat],
         color='gray',
         linestyle='--',
         marker = 'o',
         transform=ccrs.PlateCarree(),
         )

plt.plot([beij_lon, gz_lon], [beij_lat, gz_lat],
         color='r',
         linestyle='--',
         marker = 'o',
         transform=ccrs.PlateCarree(),
         )

plt.text(ny_lon - 3, ny_lat - 3, '纽约',fontsize=20,color='b',
         horizontalalignment='right',
         transform=ccrs.Geodetic())
plt.text(delhi_lon + 3, delhi_lat - 3, '新德里',fontsize=20,color='g',
         horizontalalignment='left',
         transform=ccrs.Geodetic())
plt.text(beij_lon + 3, beij_lat - 3, '北京',fontsize=20,color='r',
         horizontalalignment='left',
         transform=ccrs.Geodetic())
plt.text(gz_lon + 3, gz_lat - 3, '广州',fontsize=20,color='r',
         horizontalalignment='left',
         transform=ccrs.Geodetic())


plt.show()
````

````python
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import cartopy.feature as cfeature
from matplotlib.offsetbox import AnchoredText


def main():
    fig = plt.figure(figsize=(20,10))
    ax = plt.axes(projection=ccrs.PlateCarree())
    ax.set_extent([10, 170, -45, 30])

    # Put a background image on for nice sea rendering.
    ax.stock_img()

    # Create a feature for States/Admin 1 regions at 1:50m from Natural Earth
    states_provinces = cfeature.NaturalEarthFeature(
        category='cultural',
        name='admin_1_states_provinces_lines',
        scale='50m',
        facecolor='none')

    SOURCE = 'Natural Earth'
    LICENSE = 'public domain'

    ax.add_feature(cfeature.LAND)
    ax.add_feature(cfeature.COASTLINE)
    ax.add_feature(states_provinces, edgecolor='gray')

    # Add a text annotation for the license information to the
    # the bottom right corner.
    text = AnchoredText(r'$\mathcircled{{c}}$ {}; license: {}'
                        ''.format(SOURCE, LICENSE),
                        loc=4, prop={'size': 12}, frameon=True)
    ax.add_artist(text)

    plt.show()


if __name__ == '__main__':
    main()
````
